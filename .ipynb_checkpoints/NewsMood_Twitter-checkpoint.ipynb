```python
# Dependencies Set up
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import tweepy
import time
#import seaborn as sns
from datetime import datetime as dt
```

```python
 # Import and Initialize Sentiment Analyzer.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Import Twitter API Keys 
from config import consumer_key, consumer_secret, access_token, access_token_secret
```

```python
# Setup Tweepy API Authentication.
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# List of Target Users
#target_users = ["BBC", "CBS", "CNN", "Fox", "New York Times"]
#target_users = ("@BBC", "@CBS", "@CNN", "@FoxNews", "@NYTimes")
#target_users = ("BBC", "CBS", "CNN", "FoxNews", "NYTimes")
target_users =["@BBCNews", "@CBSNews", "@CNN","@FoxNews", "@nytimes"]
```

```python
# Set Sentiment Analysis Variables
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

# Set Variables for DataFrame
news_media = []
text = []
date = []
tweets_ago = []
```

```python
# Make a list containing the target news sources
# Looping through all the users
for target in target_users:
    
    # Initialise Tweet count
    tweet_count = 0

    # Get the last 100 tweets
    public_tweets = api.user_timeline(target, count = 100)
        # Loop through all the tweets
    for tweet in public_tweets:
        
            # Incrementing tweet_count
            tweet_count += 1
    
            # Adding values to list
            news_media.append(target)
            text.append(tweet['text'])
            date.append(tweet['created_at'])
            tweets_ago.append(tweet_count)
    
            # Run Vader Analysis on each tweet
            compound = analyzer.polarity_scores(tweet["text"])["compound"]
            pos = analyzer.polarity_scores(tweet["text"])["pos"]
            neu = analyzer.polarity_scores(tweet["text"])["neu"]
            neg = analyzer.polarity_scores(tweet["text"])["neg"]
            
            # Add each value to the appropriate list
            compound_list.append(compound)
            positive_list.append(pos)
            neutral_list.append(neu)
            negative_list.append(neg)
```

```python
# Creating the DataFrame
news_mood = pd.DataFrame({"Media Source": news_media, 
                          "Tweet": text, 
                          "Date & Time (UTC)": date,
                          "Compound Score": compound_list,
                          "Positive Score": positive_list,
                          "Neutral Score": neutral_list, 
                          "Negative Score": negative_list,
                          "Tweets Ago": tweets_ago})



# Visualize the DataFrame
news_mood.head()
```

```python
# Export the DataFrame to csv.
news_mood.to_csv("News Mood_Twitter.csv", index=False)
```

```python
# Lookup & Grab each Media Source specific data -["BBC", "CBS", "CNN", "Fox", "New York Times"]
#target_users = ("@BBC", "@CBS", "@CNN", "@FoxNews", "@NYTimes")
bbc = news_mood.loc[news_mood["Media Source"] == "@BBCNews"]
cbs = news_mood.loc[news_mood["Media Source"] == "@CBSNews"]
cnn = news_mood.loc[news_mood["Media Source"] == "@CNN"]
fox = news_mood.loc[news_mood["Media Source"] == "@FoxNews"]
nyt = news_mood.loc[news_mood["Media Source"] == "@nytimes"]
```

```python
# Set Date format for the title
Date = dt.now().strftime("(%m/%d/%Y)")
plt.figure(figsize=(10,8))
# Set the x,y limits & figure size
plt.xlim(101, -1)
#plt.ylim(-1.0, 1.0)
#plt.figsize=(16,10)

# Create scatter plots
plt.scatter(bbc['Tweets Ago'], bbc['Compound Score'], color ='skyblue', s=100, alpha=0.95,edgecolor="black", label='BBC')
plt.scatter(cbs['Tweets Ago'], cbs['Compound Score'], color ='green', s=100, alpha=0.95,edgecolor="black", label='CBS')
plt.scatter(cnn['Tweets Ago'], cnn['Compound Score'], color ='red', s=100, alpha=0.95,edgecolor="black", label='CNN')
plt.scatter(fox['Tweets Ago'], fox['Compound Score'], color ='blue',s=100, alpha=0.95,edgecolor="black", label='Fox')
plt.scatter(nyt['Tweets Ago'], nyt['Compound Score'], color ='gold', s=100, alpha=0.95,edgecolor="black", label='New York Times')

# Add legend
plt.legend(title="Media Sources",frameon = True,edgecolor = 'black',bbox_to_anchor = (1,1),fontsize=13)

# Set Grid
#plt.grid=True
plt.xlim(101, -1)
#plt.xticks(np.arange(100, 0, 20))
plt.yticks(np.arange(-1.0, 1.5, 0.5))

# Add title, x axis label, and y axis label.
plt.title("Sentiment Analysis of Media Tweets  "  +  str(Date), weight='bold').set_fontsize('13')
plt.xlabel("Tweets Ago", fontsize=13)
plt.ylabel("Tweet Polarity",fontsize=13)
# Save the plot
plt.savefig("Sentiment Analysis of Media Tweets.png", bbox_inches="tight")
# Display the visualisation
plt.show()
```
